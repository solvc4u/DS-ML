# Edges and Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('lucifer.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(image)
```

```python
g = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
edge = cv2.Canny(g, 60, 180)

fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(edge, cmap='Greys')
```

```python
contours = cv2.findContours(edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(image, contours[0], -1, (0,0,255), thickness = 2)

fig, ax = plt.subplots(1, figsize=(12,8))
plt.imshow(image)
```
