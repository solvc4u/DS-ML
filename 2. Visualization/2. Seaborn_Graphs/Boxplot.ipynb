# Boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Example 1

```python
sns.set(style="ticks", palette="pastel")

tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", hue="smoker", palette=["r", "g"],data=tips)
sns.despine(offset=10, trim=True)
```

## Example 2

```python
df = sns.load_dataset('iris')
df.head()

sns.boxplot( y=df["species"], x=df["sepal_length"], width=0.5);
plt.show()
```
