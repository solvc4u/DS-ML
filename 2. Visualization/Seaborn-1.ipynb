# Seaborn-1

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pstore = pd.read_csv("googleplaystore.csv")
pstore.head(10)
```

```python
plt.style.use("dark_background")
sns.distplot(pstore.Rating, bins=100, color="g")
plt.title("Distribution of app ratings", fontsize=15)
plt.show()
```

```python
plt.style.use("ggplot")
plt.figure(figsize=[9,7])
pstore['Content Rating'].value_counts().plot.pie()
plt.show()
```

```python
plt.style.use("seaborn-deep")
plt.figure(figsize=[9,7])
pstore['Content Rating'].value_counts().plot.barh()
plt.show()
```

```python
plt.style.use("seaborn-pastel")
plt.scatter(pstore.Size, pstore.Rating)
plt.show()
```
