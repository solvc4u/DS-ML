```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston_data = load_boston()
```

```python
boston = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
boston.head()
```

```python
boston['MEDV'] = boston_data.target
boston.head()
```

```python
boston.isnull().sum()
```

```python
boston.shape
```

```python
boston.describe()
```

```python
boston.hist(column='RM', bins=20, color='green')
plt.show()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.distplot(boston['MEDV'], bins=30)
plt.show()
```

```python
corr_matrix = boston.corr().round(2)

sns.heatmap(data=corr_matrix, annot=True)
```

```python
boston.plot(kind='scatter', x='RM', y='MEDV', figsize=(8, 6))
```

```python
boston.plot(kind='scatter', x='LSTAT', y='MEDV', figsize=(8, 6))
```

```python
X = boston[['RM']]
Y = boston[['MEDV']]
print(X.shape)
print(Y.shape)
```

```python
model = LinearRegression()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
model.fit(X_train, Y_train)
```

```python
model.intercept_.round(2)
```

```python
model.coef_.round(2)
```

```python
new_RM = np.array([6.5]).reshape(-1, 1)
model.predict(new_RM)
```

```python
y_test_predicted = model.predict(X_test)
print(y_test_predicted.shape)
type(y_test_predicted)
```

```python
plt.scatter(X_test, Y_test, label='Testing', color='r')
plt.plot(X_test, y_test_predicted, label='Prediction', linewidth=3, color='g')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show()
```

```python
res = Y_test - y_test_predicted
print(res)
```

```python
plt.scatter(X_test, res, color='g')
plt.hlines(y=0, xmin=X_test.min(), xmax=X_test.max(), linestyle='--', color='r')
plt.xlim((4,9))
plt.xlabel('RM')
plt.ylabel('Residuals')
plt.show()
```
